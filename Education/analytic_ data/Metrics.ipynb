### Задание 3

В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

Часть 1

* Доля пользователей приложения, которые принимают участие в тематическом событии. (Event Participants/MAU: Повзолит оценить долю заинтересованных пользователей и в целом интерес к данному типу активности. В знаменателе предложено учитывать MAU, поскольку пользователей не игравших больше месяца можно считать отвалившимися)  

* Доля участиков события, которые прошли все уровни. (Success Participants/Event Participants)

* Конверсии (conversions) продвижения по вехам уровней. (Данная воронка позволит определить на каком из уровней осуществляется наибольшее откалывание пользователей. Возможно, проблема в том, что уровень слишком сложный или скучный?)

* Среднее время затрачиваемое на прохождение каждого уровня. (Возможно, время затрачиваемое пользователями на прохождение уровней намного ниже предложенного ограничения, что в свою очередь может быть причиной ощущения легкодоступности предлагаемых уникальных предметов и исчезновения внутренного чувства эксклюзивности у пользователя (!предположение))

Проведение тематических событий можно расматривать как один из инструментов, напрвленых на приобретение лояльных (потенциально платящих) пользователей и в целом на удержание и вовлечение, поэтому имеет смысл оценить группу пользователей принявших участие в событии спустя время и отследить:

* Какой процент пользователей исследуемой группы принял участие в следующем тематическом событии. (Last Event Participants/Current Event Participants)

* Retention для данной группы пользователей. (Позволит сделать выводы о том, поспособствовало ли данное событие большей вовлеченности пользователей)
* Среднее время проводимое в игре пользователями данной группы до и после события. (Отдельно выделить и сравнить время группы пользователей, которые успешно прошли все уровни события)
* Средний чек пользователей данной группы до и после события. 

Часть 2

При изменении механики проведения события, может быть использован тот же набор метрик, что и для первоначальной версии события, (если мы анализируем ее успешность). Но, чтобы сделать заключение о том, насколько данный тип событий привлекателен для пользователей стоит добавить:
* Долю пользователей, которые возвращаются к игре после того, как их откатили. (Количество пользователей не покинувших событие после первого отката/Общее количество пользователей, который откатили)
* Среднее количество попыток, которое пользователи затрачивают на прохождение одних и тех же уровней, прежде чем они покинут событие (т.е. уровень настойчивости: Сумма попыток пользователей, которые в итоге сдаются/Количество этих пользователей)
